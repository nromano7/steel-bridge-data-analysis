The data in the demands and ratings files is stored as text so it needs to be cleaned using the clean_sequence method. First, load the demands file. (***NOTE: Only for 2 and 3 span***)

```python
demands = ds.load_dataset('data/demands_2.xlsx', filetype='xls', index_first_col=True)
demands.head()
```

Next, use the clean_sequence methods to convert the string sequence of values to a list of numeric values.

```python
demands = ds.clean_sequence(demands)
demands.head()
```

If you want to clean one sample at a time, use the clean method...

```python
ratings = ds.load_dataset('data/ratings_3.xlsx', filetype='xls', index_first_col=True)
ds.clean(ratings['slg_st1_pos'].values[0])
```
