# Import data and pre-process.
Steps:
1. Import raw data.
2. Inspect raw data
3. Convert string elements of dataframes to list of floats using clean_sequence method
4. Save .pickle file to store dataframes for later use.
    

```python
import dataset as ds
data = ds.load_all_data('./data')

# fields for reference
fields = data['fields']

# 1, 2, and 3-span bridge configuration and design parameters
par1 = data['parameters_1']
par2 = data['parameters_2']
par3 = data['parameters_3']

# 1, 2, and 3-span slg and fe demands
dem1 = data['demands_1']
dem2 = data['demands_2']
dem3 = data['demands_3']

# 1, 2, and 3-span slg and fe ratings
rtg1 = data['ratings_1']
rtg2 = data['ratings_2']
rtg3 = data['ratings_3']
```

Have convert dataframe string objects to lists of float values for all demands and ratings dataframes.

```python
dem1 = ds.clean_sequence(dem1)
dem2 = ds.clean_sequence(dem2)
dem3 = ds.clean_sequence(dem3)

rtg1 = ds.clean_sequence(rtg1)
rtg2 = ds.clean_sequence(rtg2)
rtg3 = ds.clean_sequence(rtg3)
```

Save each dataframe to .pickle file to /data directory for later use using to_pickle mehtod of a pandas dataframe.

```python
par1.to_pickle('./data/par1.pickle')
par2.to_pickle('./data/par2.pickle')
par3.to_pickle('./data/par3.pickle')

dem1.to_pickle('./data/dem1.pickle')
dem2.to_pickle('./data/dem2.pickle')
dem3.to_pickle('./data/dem3.pickle')

rtg1.to_pickle('./data/rtg1.pickle')
rtg2.to_pickle('./data/rtg2.pickle')
rtg3.to_pickle('./data/rtg3.pickle')

fields.to_pickle('./data/fields.pickle')
```
