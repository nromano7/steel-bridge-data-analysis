Compare slg shear to fe shear for zero_idx samples

```python
# exterior girder slg dead load shear
ext_slg_max_dlv = np.floor([slg_max_dl1v.values[x][1]+slg_max_dl2v.values[x][1] for x in range(0,5)])
ext_slg_max_dlv
```

```python
# exterior girder slg dead load shear
ext_fe_max_dlv = np.floor([np.max(fe_max_dl1v.values[x])+np.max(fe_max_dl2v.values[x]) for x in range(0,5)])
ext_fe_max_dlv
```

```python
# ratio of slg to fe dead load shear for zero_idx
ratio_ext_dlv = ext_slg_max_dlv/ext_fe_max_dlv
ratio_ext_dlv
```

```python
# exterior girder slg live load shear
ext_slg_max_llv = np.floor([slg_max_llv.values[x][1] for x in range(0,5)])
ext_slg_max_llv
```

```python
# exterior girder fe live load shear
ext_fe_max_llv = np.floor([np.max(fe_max_llv.values[x]) for x in range(0,5)])
ext_fe_max_llv
```

```python
# ratio of slg to fe live load shear for zero_idx
ratio_ext_llv = ext_slg_max_llv/ext_fe_max_llv
ratio_ext_llv
```
